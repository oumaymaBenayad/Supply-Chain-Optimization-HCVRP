# Génération des données

```python
import pandas as pd
import numpy as np
```

```python
num_points=100
coords_x=np.random.uniform(33.52794,33.58579, num_points)
coords_y=np.random.uniform(-7.64733,-7.50425,num_points)
print(coords_x)
print(coords_y)
```

```python
db_x = pd.DataFrame(coords_x, columns = ['Latitude'])
print(db_x)
```

```python
db_y = pd.DataFrame(coords_y, columns = ['Longitude'])
print(db_y)
```

```python
db=pd.concat([db_x, db_y], axis=1)
print(db)
```

```python
# extraire les données en fichier excel
db.to_excel("coordonnees_2.xlsx")
```

```python
# situer le dépôt au centre des points de vente
x0=np.average(coords_x)
y0=np.average(coords_y)
```
